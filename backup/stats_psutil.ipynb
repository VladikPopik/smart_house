```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv("/home/vladikpopik/my_projects/smart_house/backend/smart_house/data.csv")
```

```python
data
```

```python
data["timestamp"] = [datetime.datetime.fromtimestamp(x) for x in data["timestamp"]]
```

```python
data["cpu"] = [x + y + z+ d / 4 for x, y, z, d in zip(data["cpu0"], data["cpu1"], data["cpu2"], data["cpu3"])]
```

```python
print("Максимальное значение загруженности CPU = ", max(data["cpu"]), "Минимальное значение загруженности CPU = ", min(data["cpu"]))
```

```python
print("Максимальное значение загруженности по памяти = ", max(data["memory_used"]))
print("Максимальное значение загруженности по памяти в процентном отношении = ", max(data["memory_percent"]))
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 1, 1)
plt.plot(data["timestamp"], data["cpu"], label="Forecast Temp")
plt.title("Сумма нагрузки на CPU при старте системы")
plt.xlabel("Время")
plt.ylabel("CPU")
plt.xticks(rotation=45)
plt.grid(True)

```

```python
plt.subplot(1, 2, 1)
plt.plot(data["timestamp"], data["memory_used"], label="Forecast Temp")
plt.title("Сумма нагрузки на Memory при старте системы")
plt.xlabel("Время")
plt.ylabel("Memory")
plt.xticks(rotation=45)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(data["timestamp"], data["memory_percent"], label="Forecast Temp")
plt.title("Memory в процентах")
plt.xlabel("Время")
plt.ylabel("Percent")
plt.xticks(rotation=45)
plt.grid(True)
```
